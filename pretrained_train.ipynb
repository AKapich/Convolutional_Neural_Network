testing dropout

```python
test_loader = load_data('./data/test', batch_size=128, shuffle=True, transform=pretrained_transform(), num_workers=1)
train_loader = load_data('./data/train', batch_size=128, shuffle=True, transform=pretrained_transform(), num_workers=1)
valid_loader = load_data('./data/valid', batch_size=128, shuffle=True, transform=pretrained_transform(), num_workers=1)
```

```python
torch.set_num_threads(12)
for dropout in [0.2, 0.7]: # 0.5 was tested above
    vgg16 = VGG16Pretrained(num_classes=10, device=device, dropout=dropout)
    dropout_str = f"{dropout}".replace("0.", "")
    trainer = ModelTrainer(
        model=vgg16,
        train_loader=train_loader,
        device=device,
        optimizer_type='adam',  
        learning_rate=0.001,
        log_file='training_log_vgg_pretrained.json',  
        save_dir=f'./saved_models/vgg_pretrained_dropout_{dropout_str}',
        max_batches = None
    )
    trainer.train(epochs=5)
```

```python
torch.set_num_threads(12)
for w in [0.0001, 0.001, 0.01] : # 0 was tested above
    vgg16 = VGG16Pretrained(num_classes=10, device=device) 
    w_str = f"{w}".replace("0.", "")
    trainer = ModelTrainer(
        model=vgg16,
        train_loader=train_loader,
        valid_loader=valid_loader,
        device=device,
        optimizer_type='adam', 
        learning_rate=0.001,
        optimizer_params={'weight_decay': w},
        log_file=f'training_log_vgg16_weightdecay_{w_str}.json',
        save_dir=f'./saved_models/vgg16_weightdecay_{w_str}',
        max_batches=1
    )

    trainer.train(epochs=1)
```
